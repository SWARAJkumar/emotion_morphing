```python
import librosa
import numpy 
import scipy.io.wavfile as wavfile
from scipy.fftpack import dct
import matplotlib.pyplot as plt
from scipy import signal as sig
import python_speech_features as psf
from matplotlib import cm
import os
from IPython.display import Image
```

```python
wav_file="decoder_anger.wav"
sample_rate, signal = wavfile.read(wav_file)  # File assumed to be in the same directory
signal = signal[0:int(2 * sample_rate)]


Time=numpy.linspace(0, len(signal)/sample_rate, num=len(signal))

plt.plot(Time,signal)
plt.show()
```

```python
mfccs=librosa.feature.mfcc(y=signal, sr=22050, S=None, n_mfcc=20)
print(mfccs.shape)
import librosa.display
librosa.display.specshow(mfccs, x_axis='time', cmap='gist_heat')

plt.show()
```

```python
def graph_spectrogram(wav_file):
    rate, data = get_wav_info(wav_file)
    nfft = 512#length of the windowing segments
    fs = rate  # Sampling frequency

    
    plt.axis('off')
    file_name= wav_file[:-4]+".png"
    fig = plt.figure(frameon=False)
    fig.set_size_inches(2,2)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    pxx, freqs, bins, im = plt.specgram(data, nfft,fs,noverlap=256, cmap="plasma")
    #plt.imshow(mag_frames, interpolation='nearest', cmap=cm.seismic, origin='lower', aspect='auto')
    fig.savefig(file_name, dpi=512)

     
   # plt.savefig(file_name,
                #dpi=300, # Dots per inch
                #frameon='false',
                #aspect='normal',
                #bbox_inches='tight',
                #pad_inches=0
       #        ) # Spectrogram saved as a .png 
    
    plt.show()

def get_wav_info(wav_file):
    rate, data = wavfile.read(wav_file)
    return rate, data


wav_file = 'YAF_late_disgust.wav' # Filename of the wav file
graph_spectrogram(wav_file)
```
